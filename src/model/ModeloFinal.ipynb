Importamos las librerías necesarias

```python
import xgboost
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

### Preparamos el target 

```python
df = pd.read_csv('/Users/alexcar934/Desktop/src/data/processed/Datos.csv')
values = []
for i in df['Valor de Mercado']:
    if i < 1750000:
        values.append(0)
    elif i >= 1750000 and i < 5000000:
        values.append(1)
    elif i >= 5000000 and i < 15000000:
        values.append(2)
    elif i >= 15000000:
        values.append(3)
df['Valor de Mercado'] = values
df = df.iloc[:,1:]
df
```

```python
X = df.drop(['Nombre del jugador','Valor de Mercado'], axis=1)
y = df['Valor de Mercado']
```
