# Problema serio

## Hay anomalías en los datos:

En la columna 'Minutos Jugados', hay ciertos jugadores de los que se ha extraído que han jugado 

(Por Ejemplo):

    Jan Oblak, portero titular del Atlético de Madrid, marca un registro de 333 minutos jugados cuando ha jugado 3330.
    Hay jugadores que registran 31 minutos cuando han jugado 3100.


## Solución propuesta:

Multiplicar por 100 los valores de la columna Minutos jugados: 

    Los jugadores que hayan jugado menos de 35 minutos
    SOLO si han encajado un valor igual o superior a 15 goles.

* En el caso de haber jugado 34 o 340 minutos, no has podido encajar 15 goles (en ese caso eres un outlier).
* Los valores que sean menores de 34 se multiplican por 100, por lo que pasan a haber jugado --> valores menores de 3400 minutos 
(tiempo suficiente para haber encajado 16 goles)

Multiplicar por 10 los valores de la columna Minutos jugados:

    Los jugadores que hayan jugado menos de 343 minutos
    SOLO si han encajado un valor igual o superior a 15 goles.

* En el caso de haber jugado 340 minutos, no has podido encajar 15 goles (en ese caso, eres un outlier)
* Los valores que sean menores de 343 se multiplican por 10, por lo que pasan a haber jugado --> valores menores de 3430 minutos 
(tiempo suficiente para haber encajado 16 goles).

* EXPLICACIÓN DE LOS VALORES:

Estas ligas las comprenden:
* Máximo 20 equipos
* Juegan entre ellos dos veces --> 38 partidos como máximo por equipo
* Los partidos son de 90 minutos --> un total de 3420 minutos por temporada.

* Que te metan 15 goles en 3 partidos y medio es complicado (342 minutos), en 3420 minutos... ¡No tanto!

```python
minutos = []
for i in df.index:
    if np.all(df['Minutos jugados'][i] <= 34) and np.all(df['Goles encajados jugando'][i]>=15):
        minutos.append(df['Minutos jugados'][i]*100)
    elif np.all(df['Minutos jugados'][i]< 343) and np.all(df['Goles encajados jugando'][i]>=15):
        minutos.append(df['Minutos jugados'][i]*10)
    else:
        minutos.append(df['Minutos jugados'][i]) 

df['Minutos jugados'] = minutos
df.sort_values(by='Minutos jugados').head()
```
